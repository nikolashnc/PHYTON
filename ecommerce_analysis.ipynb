```python
#Calculating Sales
data['Sales'] = data['Quantity'] * data['UnitPrice']
data
```

```python
#Insight 1: Monthly Revenue Trend
#Question: How does the monthly revenue trend look over the entire period?

# Calculating monthly revenue
monthly_revenue = data.groupby('Year_Month')['Revenue'].sum()

# Plotting monthly revenue trend
plt.figure(figsize=(8, 5))
monthly_revenue.plot(kind='line', marker='o', color='blue')
plt.title('Monthly Revenue Trend')
plt.xlabel('Year_Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Insight 2: Products with the most revenue.

#Calculating Products with the most revenue.
most_revenue_product= data.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)

# Plotting top-selling products
plt.figure(figsize=(10, 6))
most_revenue_product.plot(kind='bar', color='red')
plt.title('Products with the most revenue')
plt.xlabel('Product Description')
plt.ylabel('Revenue')
plt.xticks(rotation=80)
plt.tight_layout()
plt.show()
```

```python
#Insight 3: Top Selling Products 

# Calculating quantity sold per product
product_quantity_sold = data.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

# Plotting top-selling products
plt.figure(figsize=(10, 6))
product_quantity_sold.plot(kind='bar', color='blue')
plt.title('Top 10 Selling Products')
plt.xlabel('Product Description')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=80)
plt.tight_layout()
plt.show()
```

```python
#Insight 4: Sales distribution by country 

# Calculating sales distribution by country
country_sales = data.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(5)

# Plotting sales distribution by country
plt.figure(figsize=(6, 6))
country_sales.plot(kind='bar', color='purple')
plt.title('Sales Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Insight 5: Daily sales variation 

# Extracting day of the week from 'InvoiceDate'
data['Day_of_Week'] = data['InvoiceDate'].dt.day_name()

# Calculating daily sales variation
daily_sales = data.groupby('Day_of_Week')['Revenue'].sum().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plotting daily sales variation
plt.figure(figsize=(6, 6))
daily_sales.plot(kind='bar', color='red')
plt.title('Daily Sales Variation')
plt.xlabel('Day of the Week')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```

```python
#Insight 6: Unit Price Distribution

#Plotting Unit price frequency through a histogram
plt.figure(figsize=(8, 6))
plt.hist(data['UnitPrice'], bins=100, color='lightseagreen', range=(0, 100))  # Adjust the range accordingly
plt.title('Unit Price Distribution')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()

```
